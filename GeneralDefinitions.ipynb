# Probability Dictionary
$A\cup B$: event if and only if at least one of $A$ and $B$ occurs. \
$A\cap B$: event if and only if both $A$ and $B$ occur. \
$A^C$ or $B^C$: event if and only if event $A$ (or $B$) does not occur.

| English | Set |
| -- | -- |
| A or B | $A$ $\cup$ $B$ |
| A and B | $A$ $\cap$ $B$ | 
| Not A | $A^c$ |
| A or B, but not both | $(A$ $\cap$ $B^c$) $\cup$ $(A^c$ $\cap$ $B)$ |
| At least on of $A_1,...,A_n$ | $A_1$ $\cup$ ... $\cup$ $A_n$ or $\cup_{j=1}^{n}A_j$ |
| All of $A_1,..., A_n$ | $A_1$$\cap$ ... $\cap$ $A_n$ or $\cap _{j=1}^{n} A_j$ |
| A implies B | $A$$\subseteq$$B$ | 
| A and B are mutually exlc. | $A$$\cap$$B$ = $\emptyset$ | 
| $A_1,..., A_n$ are a partition of S | $A_1$$\cup$ ... $\cup$ $A_n$ = $S$, $A_i$$\cap$ $A_j$ = $\emptyset$ for $i \neq j$ | 

## De Morgan's Laws:
\begin{equation}
  (A\cup B)^c=A^c\cap B^c \text{ and } (A\cap B)^c=A^c\cup B^c 
\end{equation}
